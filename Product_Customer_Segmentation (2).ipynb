# **Introduction**

In marketing and selling products or services, it is essential to put in mind that different customers have different preferences, needs, and behaviors, and it's crucial to understand these differences to effectively reach and engage with them. One powerful way to do this is by segmenting customers by age. By doing so, you can tailor your marketing strategies to better resonate with each group and ultimately drive more sales and customer loyalty.

Moreover a company that generates high profits is more valuable than a company that generates low profits. The value of a business is determined by its ability to generate future profits. Hence, it is vital for a company to understand which product is generating high profit.

## **Analysis Objective**
This analysis is intended to find out the
* The effects of different age group on the sales revenue and profit
* The product that is generating the highest revenue and Profit

## How is the Customer Gender ratio?

```python
Gender=df['Customer_Gender'].value_counts()
#create a pie chart
plt.pie(Gender.values, labels=Gender.index, colors=["Teal", "pink"], autopct='%1.1f%%')
plt.title("Gender Ratio")
#ADD A CIRCLE AT THE CENTRE TO TRANSFORM IT TO DONUT CHART
my_circle=plt.Circle((0,0), 0.4, color='white')
p=plt.gcf()
p.gca().add_artist(my_circle)
plt.show()
```

```python
df["Customer_Gender"].unique()
```

* We note that the ratio of male customer is higher compared to that of the female customers.

## How is the overall Age_Group distribution

```python
Age_Group=df['Age_Group'].value_counts()
Age_Group
plt.pie(Age_Group.values, labels=Age_Group.index, colors=["teal", "aqua", "green", "red"], autopct='%1.1f%%')
plt.title('Age Group Distribution')
plt.show()
```

* The age group "Adult(35-64)" has the highest distribution of 49.4% and the "Seniors(64+)" being the least with distribution of 0.6%

## Which Age_Group is generating the highest profit?

```python
sns.barplot(df, x= df["Profit"],  hue= df["Age_Group"], estimator=sum, palette=["teal", "aqua", "green", "red"])
plt.title("Profit by Age_Group")
```

* The overall highest profit is generated by the Age_Group Adults(35-64)
* This can be due to the fact that this age group has the highest distribution.

## Which product has the highest quantity order in each age group?

```python
sns.countplot(df, x= df["Age_Group"],  hue= df["Product_Category"], palette=["teal", "aqua", "red"])
plt.title("Product order quantity by Age Group")
```

* Accessories is seen to have the highest quantity order across all age groups.
* Bike takes the second position in orders accross all age groups
* Clothes are the least in order
* It can be noted that among the Youth(<25), accessories and clothes almost equal ratio of order


## How is the Age_Group distributed in each country?

```python
sns.countplot(df, x= df["Country"],  hue= df["Age_Group"], palette=["teal", "aqua", "green", "red"])
plt.title("Age_Group Distribution by Country")
```

* From the analysis, it is clear that in every country represented here, the Age_Group Adults(35-64) has the highest number and the Seniors(64+) is the least.
* United states has the highest number of each age_group compared to other countries, with the highest of it all being the adults(35-64)

## Which Country generated the highest revenue?

```python
sns.barplot(x="Country", y="Revenue", data=df, estimator=sum, color="teal")
plt.title("Revenue by Country")
plt.xlabel("Country")
plt.ylabel("Revenue")
plt.show()
```

* From the visualization, United States is seen to generate the highest revenue while Cananda generates the lowest revenue.

## How is the product Profit by Country?

```python
sns.barplot(df, x= df["Country"], y=df["Profit"], hue= df["Product_Category"], estimator=sum, palette=["teal", "red", "green"])
plt.title("Product profit by ")
```

* Bike is seen to be generating the highest profit across all countries followed by accessories and lastly clothing.
* It is also notable that in United States all products are generating higher profit as compared to other products in other countries.


## Does the Product Category with the higher Order Quantity have the higher Profit?

```python
sns.scatterplot(x="Order_Quantity", y="Profit", hue="Product_Category", palette=["teal", "red", "green"], data=df)
plt.title("Profit per Product Order Quantity")
plt.xlabel('Order Quantity')
plt.ylabel('Profit')
plt.show()
```

* From this analysis we can see that the bike has the lowest Order Quantity while the Accessories and the Clothes has higher Order Quantity.
*However it is notable that it is not the product with the highest Order Quantity that generates the highest profit as we can see the bike with the lowest Order Quantity is generating the highest Profit

## Which product made the highest profit?

```python
sns.barplot(x = "Product_Category", y="Profit", hue="Product_Category", legend=False, palette=["teal", "red", "aqua"], data = df)
plt.title(" Product Category Vs Profit ")
plt.xlabel("Product Category")
plt.ylabel("Profit")
```

* Bikes made the highest profit and accessories made the least profit

## How is the profit margin per given product category?

```python
category_column="Product_Category"
numerical_column="Profit"

sns.boxplot(x=category_column, y=numerical_column, hue="Product_Category", legend=False, data=df, palette=["teal", "red", "aqua"])

plt.show()
```

* Procuct catagory Bike is seen to have the highest profit margin




## **Findings**
* The adults (35-64) age group is highly distributed across all country and seniors (64+) is the least distributed.
* Profit generation rate varies in different age groups with the highest profit being generated by the age group adults (35-64)
* United States has the distribution of all the age groups
* United States is generating the highest revenue and profit which can be associated by the highest distribution of all age groups in the country.
* The product category with the highest order quantity does not translate to be the one with the higher sum profit as it is observed that the bike category has the lowest order quantity, however it has the highest sum profit which can be associated by high profit margin of the product bike.
